```python
import qiskit
import numpy as np
```

```python
from qiskit import Aer, execute, QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Statevector
```

```python
simulator = Aer.get_backend('statevector_simulator')
```

```python
n=4
r=4
qc = QuantumCircuit()
qr_c = [ QuantumRegister(r, 'qr_{}'.format(i)) for i in range(n) ]
for i in range(n):
    qc.add_register(qr_c[i])
    qc.ry(2*np.arccos(1/np.sqrt(r)), qr_c[i][0])
    for j in range(r-1):
        if j == r-2:
            qc.cx(qr_c[i][j],qr_c[i][j+1])
        else:
            qc.cry(2*np.arccos(1/np.sqrt(r-j-1)), qr_c[i][j], qr_c[i][j+1])
    for k in range(1,r-1):
        j=r-k-1
        qc.cx(qr_c[i][j-1], qr_c[i][j])
    qc.x(qr_c[i][0])
print(qc.draw(fold=150))

latest_replacement = (1, -11)
i, j = np.abs(latest_replacement)
qc = QuantumCircuit()
qr_c = [ QuantumRegister(r, 'qr_{}'.format(k)) if k!=j//r else QuantumRegister(r-1, 'qr_{}'.format(k)) for k in range(n) ] #Remove 1 qubit replacement
for k in range(n):
    #We remove x_j as follows, where j = latest_replacement[1]
    qc.add_register(qr_c[k])
    if k!=j//r: #Initialise as before
        #Initialise a complete W-state on this block as it is unaffected by the replacement
        qc.ry(2*np.arccos(1/np.sqrt(r)), qr_c[k][0])
        for l in range(r-1):
            if l == r-2:
                qc.cx(qr_c[k][l],qr_c[k][l+1])
            else:
                qc.cry(2*np.arccos(1/np.sqrt(r-l-1)), qr_c[k][l], qr_c[k][l+1])
        for l in range(1,r-1):
            m=r-l-1
            qc.cx(qr_c[k][m-1], qr_c[k][m])
        qc.x(qr_c[k][0])
    elif k==j//r and i//r != j//r: #Now for the block with removed qubit -> if the replacements involved qubits from different blocks, W-state removed final qubit:
        qc.ry(2*np.arccos(1/np.sqrt(r)), qr_c[k][0])
        for l in range(r-2):
            qc.cry(2*np.arccos(1/np.sqrt(r-l-1)), qr_c[k][l], qr_c[k][l+1])
        for l in range(1,r-1):
            m=r-l-1
            qc.cx(qr_c[k][m-1], qr_c[k][m])
        qc.x(qr_c[k][0])
    elif k==j//r and i//r == j//r and latest_replacement[1] < 0: 
        #If replacement is within same car_block and z_1 = -z_j
        #We remove x_j, then everything except x_i must be 0 since one of x_i or x_j must be 1 according do z_i = -z_j.
        qc.h(qr_c[i//r][i%r])
    else: 
        #if  the replacement occur on the same block, and if z_j = z_j
        #Remove x_j, then x_i must be 0, and everything else is in a W-state:
        w_regs = [qr_c[k][j] for j in range(r-1) if j != i%r]
        N = len(w_regs)
        qc.ry(2*np.arccos(1/np.sqrt(N)), w_regs[0])
        for s in range(N-1):
            if s == N-2:
                qc.cx(w_regs[s],w_regs[s+1])
            else:
                qc.cry(2*np.arccos(1/np.sqrt(N-s-1)), w_regs[s], w_regs[s+1])
        for s in range(1,N-1):
            m=N-s-1
            qc.cx(w_regs[m-1], w_regs[m])
        qc.x(w_regs[0])
print(qc.draw(fold=150))
replacement_same_sign = True
```

```python
simulator = Aer.get_backend('qasm_simulator')
counts = execute(qc, simulator, shots = 100000).result().get_counts()
```
