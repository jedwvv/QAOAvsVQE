```python
RI_probs = []
RI_approx = []
for i in range(11):
    with open('QAOAvsVQE/results_3cars3routes/RI_F_{}.csv'.format(i), 'r') as f:
        b = np.loadtxt(f, delimiter=',')
    RI_approx.append(b[0:4])
    RI_probs.append(b[4:8])
RI_approx_avg = np.mean(RI_approx, axis=0)
RI_approx_probs = np.mean(RI_probs, axis=0)
```

```python
TQA_probs = []
TQA_approx = []
for i in range(11):
    with open('QAOAvsVQE/results_3cars3routes/TQA_F_{}.csv'.format(i), 'r') as f:
        b = np.loadtxt(f, delimiter=',')
    TQA_approx.append(b[0:4])
    TQA_probs.append(b[4:8])
TQA_approx_avg = np.mean(TQA_approx, axis=0)
TQA_approx_probs = np.mean(TQA_probs, axis=0)
```

```python
plt.plot([1, 2, 3, 4], RI_approx_avg, 'b--', label = "Layer based random initialization")
plt.plot([1, 2, 3, 4], TQA_approx_avg, 'r--', label = "TQA initialization")
plt.xlabel("p")
plt.xticks([1,2,3,4])
plt.ylabel("1-r")
plt.title("1-r vs p")
plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], RI_approx_probs, 'b--', label = "Layer based random initialization")
plt.plot([1, 2, 3, 4], TQA_approx_probs, 'r--', label = "TQA initialization")
plt.xticks([1,2,3,4])
plt.xlabel("p")
plt.ylabel("probability of success")
plt.title("Probability of success vs p")
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```
